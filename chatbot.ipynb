```python
from openai import OpenAI
import httpx
import os

http_client=httpx.Client(verify=False)
client = OpenAI(
    base_url='https://genai-api-dev.dell.com/v1',
    http_client=http_client,
    api_key=os.environ["DEV_GENAI_API_KEY"]
)
```

```python
streaming = True
max_output_tokens = 10000

codestral_prompt = ''' 
You are a SQL expert. Convert the following PostgreSQL DDL statement into a valid and optimized Teradata SQL statement. Ensure all data types and distribution logic are correctly translated. Use create 
table multiset, unicode characterset and instead of distributed primary index and primarykey column
 
CREATE TABLE mkt_lz.jing_cn_cons_sms_campaign (
	cmpgn_nm varchar(1000) NULL,
	cmpgn_id varchar(800) NOT NULL,
	sms_msg text NULL,
	landing_page_url text NULL,
	"segment" varchar(400) NULL,
	src_extrc_file_nm varchar(200) NOT NULL,
	dw_src_site_id int2 NOT NULL,
	dw_ins_dtsz timestamptz NOT NULL,
	dw_upd_dtsz timestamptz NOT NULL,
	dw_ld_grp_val numeric(18, 2) NOT NULL,
	dw_etl_sess_nm varchar(200) NOT NULL,
	dw_row_hash_val varchar(32) NOT NULL
)
DISTRIBUTED RANDOMLY;'''

available_models = ["codellama-13b-instruct", "sqlcoder-7b-2", "codestral-22b-v0-1", "llama-3-sqlcoder-8b"]

model_selected = available_models[2]

models_prompts = {"codestral-22b-v0-1" : codestral_prompt
                 }

prompt_selected = models_prompts[model_selected]

print(f"Model: {model_selected}")
completion = client.completions.create(
    model=model_selected,
    
    max_tokens=max_output_tokens,
    prompt=prompt_selected,
    stream=streaming)

if streaming:
    for chunk in completion:
        print(chunk.choices[0].text, end='')
else:
    print(completion.choices[0].text)
```

```python
import re

# Function to classify SQL type
def classify_sql_type(sql_text: str) -> str:
    ddl_keywords = {'create', 'alter', 'drop', 'truncate', 'rename'}
    dml_keywords = {'insert', 'update', 'delete', 'merge', 'select'}
    tokens = set(re.findall(r'\b\w+\b', sql_text.lower()))

    if tokens & ddl_keywords:
        return 'DDL'
    elif tokens & dml_keywords:
        return 'DML'
    else:
        return 'UNKNOWN'
    
def enhance_prompt_based_on_sql_type(sql_text: str) -> str:
    sql_type = classify_sql_type(sql_text)
    
    if sql_type == 'DDL':
        prompt = (
            '''You are a SQL expert. Convert the following PostgreSQL into an optimized Teradata SQL statement.
            Preserve data types, structure, and logic as appropriate. Use create table multiset, unicode characterset and instead of distributed use primary index and primarykey column\n'''
        )
    elif sql_type == 'DML':
        prompt = (
            "You are a SQL expert. Convert the following PostgreSQL DML query into a valid and efficient Teradata SQL statement. "
            "Ensure syntax and functions are properly translated:\n"
        )
    else:
        prompt = ""
    
    return prompt + sql_text
```

```python
streaming = True
max_output_tokens = 10000

codestral_prompt = '''
 
CREATE TABLE mkt_lz.jing_cn_cons_sms_campaign (
	cmpgn_nm varchar(1000) NULL,
	cmpgn_id varchar(800) NOT NULL,
	sms_msg text NULL,
	landing_page_url text NULL,
	"segment" varchar(400) NULL,
	src_extrc_file_nm varchar(200) NOT NULL,
	dw_src_site_id int2 NOT NULL,
	dw_ins_dtsz timestamptz NOT NULL,
	dw_upd_dtsz timestamptz NOT NULL,
	dw_ld_grp_val numeric(18, 2) NOT NULL,
	dw_etl_sess_nm varchar(200) NOT NULL,
	dw_row_hash_val varchar(32) NOT NULL
)
DISTRIBUTED RANDOMLY;'''

available_models = ["codellama-13b-instruct", "sqlcoder-7b-2", "codestral-22b-v0-1", "llama-3-sqlcoder-8b"]

model_selected = available_models[2]

prompt_selected = enhance_prompt_based_on_sql_type(codestral_prompt)

print(f"Model: {model_selected}")
completion = client.completions.create(
    model=model_selected,
    
    max_tokens=max_output_tokens,
    prompt=prompt_selected,
    stream=streaming)

if streaming:
    for chunk in completion:
        print(chunk.choices[0].text, end='')
else:
    print(completion.choices[0].text)
```
